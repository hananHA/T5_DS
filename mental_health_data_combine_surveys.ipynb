## Combine Surveys

This notebook is only about combining the data of surveys from 2014 to 2020

```python
# Import needed modules
import pandas as pd
import re
```

```python
# general function to drop columns
def drop_cols(dataFrame, cols, inplace):
    dataFrame.drop(cols, axis=1, inplace=inplace)
```

```python
df2014 = pd.read_csv('mental_health_data/mental_health_in_tech_survey_2014.csv')
df2016 = pd.read_csv('mental_health_data/mental_health_in_tech_survey_2016.csv')
df2017 = pd.read_csv('mental_health_data/mental_health_in_tech_survey_2017.csv')
df2018 = pd.read_csv('mental_health_data/mental_health_in_tech_survey_2018.csv')
df2019 = pd.read_csv('mental_health_data/mental_health_in_tech_survey_2019.csv')
df2020 = pd.read_csv('mental_health_data/mental_health_in_tech_survey_2020.csv')
```

```python
df_all = df2014.append(df2016, ignore_index=True)
df_all = df_all.append(df2017, ignore_index=True)
df_all = df_all.append(df2018, ignore_index=True)
df_all = df_all.append(df2019, ignore_index=True)
df_all = df_all.append(df2020, ignore_index=True)
```

```python
df_all.head()
```

```python
for col in df_all.columns:
    print(col)
```

```python
drop_cols(df_all, ['#'], True)
for col in df_all.columns:
    new_col = re.sub('\W+',' ', col)
    new_col = new_col.replace('strong', '')
    new_col = new_col.strip()
    new_col = new_col.replace('issues', 'disorders')
    new_col = new_col.replace('issue', 'disorder')
    new_col = new_col.replace('  live', ' live')
    new_col = new_col.replace('  work', ' work')
    new_col = new_col.replace('  in', ' in')
    df_all.rename(columns = {col: new_col}, inplace=True)
```
